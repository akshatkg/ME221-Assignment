## Importing required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Defining constants

```python
A = 1.537
B = 7.421e-6
n = 8
```

## Defining Energies

```python
def Ea(xs, coeffA):
    ys = np.zeros(1)  # Initialise an array of zeros
    ys = - (coeffA / xs) # calculating attractive energy
    return ys

def Er(xs, coeffB):
    ys = np.zeros(1)  # Initialise an array of zeros
    ys = (coeffB / xs ** 8) # calculating repulsive energy
    return ys

def E(xs, coeffA, coeffB):
    return Ea(xs, coeffA) + Er(xs, coeffB) # calculating Net Energy
```

```python
xs = np.linspace(0.00001,1.4, 1000)
```

# Q1 (Energy)

```python
plt.style.use('classic')
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(xs, Ea(xs, A), '-.', label='Attractive Energy')  # Plot some data on the axes.
ax.plot(xs, Er(xs, B), '--', label='Repulsive Energy')
ax.plot(xs, E(xs, A, B), '-', label='Net Energy')
ax.plot(xs, np.zeros(1000), color= 'gray')
plt.axvline(x=0.2341, color = 'black', ymin = -10,linestyle = '-.')
plt.plot(0.2341,0, color='black', marker='o', label='Ro')
ax.set_ylim(-20, 20)
ax.set_xlabel(r'Interatomic separation (r)')  # Add an x-label to the axes.
ax.set_ylabel(r'<- Attraction                    Potential Energy (E)                Repulsion ->')  # Add a y-label to the axes.
ax.set_title("Plot 1")  # Add a title to the axes.
ax.legend()  # Add a legend
```

## Defining Forces

```python
def Fa(xs, coeffA):
    ys = np.zeros(1)  # Initialise an array of zeros of the required length.
    ys = (coeffA / xs ** 2)
    return ys

def Fr(xs, coeffB):
    ys = np.zeros(1)  # Initialise an array of zeros of the required length.
    ys = -8 * (coeffB / xs ** 9)
    return ys

def F(xs, coeffA, coeffB):
    return Fa(xs, coeffA ) + Fr(xs, coeffB)
```

# Q1 (Force)

```python
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(xs, Fa(xs, A), '-.', label='Attractive Force')  # Plot some data on the axes.
ax.plot(xs, Fr(xs, B), '--', label='Repulsive Force')
ax.plot(xs, F(xs, A, B), '-', label='Net Force')
ax.plot(xs, np.zeros(1000), '-', color= 'gray')
plt.plot(0.2341,0, color='black', marker='o', label='Ro')
ax.set_ylim(-40, 40)
ax.set_xlabel(r'Interatomic separation (r)')  # Add an x-label to the axes.
ax.set_ylabel('<- Repulsion                      Force (F)                  Attraction ->')  # Add a y-label to the axes.
ax.set_title("Plot 2")  # Add a title to the axes.
ax.legend()  # Add a legend
```
